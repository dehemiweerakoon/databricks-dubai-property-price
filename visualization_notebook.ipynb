```python
gold_type_df = spark.sql("""
    SELECT prop_sb_type_en, rooms_en, avg_price_per_sqm
    FROM gold_price_by_type
    LIMIT 10
""")

pdf = gold_type_df.toPandas()
```

```python
import matplotlib.pyplot as plt

temp_df = pdf.pivot_table(
    index="prop_sb_type_en",
    columns="rooms_en",
    values="avg_price_per_sqm",
    aggfunc="sum",
    fill_value=0
)

fig, ax = plt.subplots(figsize=(10, 6))
temp_df.plot(kind="barh", stacked=True, ax=ax, colormap="tab10")

ax.set_xlabel("(avg_price_per_sqm)")
ax.set_ylabel("prop_sb_type_en")
ax.set_title("Avg Price per Sqm by Property Sub-Type and Room Configuration")
ax.legend(title="rooms_en", bbox_to_anchor=(1.02, 1), loc="upper left")

plt.tight_layout()
```

This chart shows the average price for the sqaure meter for each type of the property that is grouped by the room type. So Hotel has higest overall room types and they constribute the price increse of that propery
--- why ? which property has high value propeties for higher commisions and to get budget friendly decisions.

```python
gold_area_df = spark.sql("""
    SELECT area_en, avg_price_per_sqm AS avg_price
    FROM gold_price_by_area
    ORDER BY avg_price DESC
    LIMIT 10
""")
pdf1 = gold_area_df.toPandas()
```

```python
import matplotlib.pyplot as plt

fig, ax = plt.subplots(figsize=(10, 6))
ax.barh(pdf1["area_en"], pdf1["avg_price"], color="#1f77b4")
ax.invert_yaxis()  
ax.set_xlabel("avg_price")
ax.set_ylabel("area_en")
ax.set_title("Top 10 Districts by Avg Price per Sqm")

plt.show()
```

This chart shows the 10 most valuable proprty consists districts in dubai using the avg pricer per the squre meters
--- why ? which district has high value propeties for higher commisions and to get budget friendly decisions.

```python
gold_trend_df = spark.sql("SELECT * FROM gold_monthly_view ORDER BY year_month")
pdf3 = gold_trend_df.toPandas()
```

```python


fig, ax1 = plt.subplots(figsize=(10, 6))

ax1.plot(pdf3["year_month"], pdf3["sum_trans_value"], marker="o", color="#1f77b4", label="Transaction Sum")
ax1.set_xlabel("year_month")
ax1.set_ylabel("Transaction Sum", color="#1f77b4")
ax1.tick_params(axis="y", labelcolor="#1f77b4")
plt.xticks(rotation=45)

ax2 = ax1.twinx() 
ax2.plot(pdf3["year_month"], pdf3["avg_price_per_sqm"], marker="s", color="#d62728", label="Avg Price/Sqm")
ax2.set_ylabel("Avg Price per Sqm", color="#d62728")
ax2.tick_params(axis="y", labelcolor="#d62728")

ax1.set_title("Monthly Transaction Sum vs. Avg Price per Sqm (2026)")
fig.tight_layout()
```
